# **MatPlotlib**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 1000)   # Create 1000 point between 0 andd 10 equally spaced.(Something to take the sin of.)
y = np.sin(x)
plt.plot(x, y)     # basic continuous line plot.
plt.scatter(x[::10], y[::10])  # plot by leaving them as individual points, dont merge as a line.(Choosing every tenth point to not make the points too dense.)
```

Multiple plots in the same graph with different colors and properties.
