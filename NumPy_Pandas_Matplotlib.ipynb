##### Search Sorted
There is a method called searchsorted() which performs a binary search in the array, and returns the index where the specified value would be inserted to maintain the search order.

```python
import numpy as np

arr = np.array([6, 7, 8, 9])
x = np.searchsorted(arr, 7)
print(x)
```

```python
import numpy as np

arr = np.array([1, 3, 5, 7])
x = np.searchsorted(arr, [2, 4, 6])
print(x)
```

The return value is an array: [1 2 3] containing the three indexes where 2, 4, 6 would be inserted in the original array to maintain the order.

#### Slicing numpy arrays

Slicing an array creates a view of the same array, accessing the same memory as the original array. 
While if we slice a normal python list with numpy, it creates a copy of it.


```python
import numpy as np

A = np.arange(0, 100)
print(A[::5])    # goes from the front side
print(A[::-5])   # goes from the end side 
```

```python
# to reverse an array
print(A[::-1])
```
